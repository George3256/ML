```python
import os
import pandas as pd

filename = os.path.join("datasets", "hockey", "new_hockey_2019_01_18.csv")
hockey = pd.read_csv(filename)
```

```python
hockey.head()
```

```python
import matplotlib.pyplot as plt

def plot_statistic_hosts(comand, column):
    global data
    #filepath = os.path.join("plotting", "Комманда_{}   Период_{}.png". format(comand, column))
    data = hockey[["DataTime", column]].where(hockey["Hosts"] == comand).dropna()
    plt.figure(figsize=(30, 5))
    plt.ylim([0,10])
    plt.xticks(rotation=90)
    plt.xlabel(column)
    plt.ylabel(comand)
    plt.plot(data["DataTime"], data[column])
    plt.xticks(rotation=90)
    #plt.savefig(filepath)
    #plt.close()
    # plt.show()

def plot_statistic_guests(comand, column):
    global data
    #filepath = os.path.join("plotting", "Комманда_{}   Период_{}.png". format(comand, column))
    data = hockey[["DataTime", column]].where(hockey["Guests"] == comand).dropna()
    plt.figure(figsize=(30,5))
    plt.xticks(rotation=90)
    plt.ylim([0,10])
    plt.xlabel(column)
    plt.ylabel(comand)
    plt.plot(data["DataTime"], data[column])
    
    #plt.savefig(filepath)
    #plt.close()
    # plt.show()

def plot_statistic_host(comand):
    list_column_hosts = ["1PHosts","2PHosts","3PHosts","OTHosts","SSHosts","CountHosts"]
    for column in list_column_hosts:
        plot_statistic_hosts(comand, column)
        
def plot_statistic_guest(comand):
    list_column_guests = ["1PGuests","2PGuests","3PGuests","OTGuests","SSGuests","CountGuests"]
    for column in list_column_guests:
        plot_statistic_hosts(comand, column)
```

```python
host = "Нью-Джерси"
guests = "Филадельфия"
```

```python
column = "CountHosts"
data = hockey[["DataTime", column ]].where(hockey["Hosts"] == host).dropna()
print(host)
print("Математическое ожидание {}, Дисперсия {} Среднеквадратическое отклонение {}".format(data.CountHosts.mean(),data.CountHosts.var(),data.CountHosts.std()))
plot_statistic_hosts(host, column)
column = "CountGuests"
data = hockey[["DataTime", "CountGuests"]].where(hockey["Guests"] == guests).dropna()
print(guests)
print("Математическое ожидание {}, Дисперсия {} Среднеквадратическое отклонение {}".format(data.CountGuests.mean(),data.CountGuests.var(),data.CountGuests.std()))
plot_statistic_guests(guests, column)
```
