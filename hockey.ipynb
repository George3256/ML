```python
import os
HOCKEY_PATH = os.path.join("datasets", "hockey")
```

```python
import pandas as pd

def load_hockey_data(hockey_path=HOCKEY_PATH):
    csv_path = os.path.join(hockey_path, "hockey.csv")
    return pd.read_csv(csv_path)
```

```python
hockey = load_hockey_data()
hockey.head()
```

```python
def count1P(serias):
    list_count = []
    j = 0
    while j < serias.count():
        current_seria = serias[j]
        count = 0
        if str(current_seria) != 'nan':
            x = str(current_seria).split(' ')
            for i in x:
                if int(i) <= 20 and int(i) != -1:
                    count += 1
        list_count.append(count)
        j += 1
    new_series = pd.Series(list_count)
    return new_series
```

```python
hockey['1PHosts'] = count1P(hockey['Время Х'])
```
